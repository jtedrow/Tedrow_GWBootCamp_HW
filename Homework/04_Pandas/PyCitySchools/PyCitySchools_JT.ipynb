```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculations
total_schools = school_data["school_name"].nunique()
total_students = student_data["student_name"].count()
total_budget = school_data["budget"].sum()
avg_math = student_data["math_score"].mean()
avg_reading = student_data["reading_score"].mean()
avg_overall = (avg_math + avg_reading) / 2
math_pass = len(student_data[student_data["math_score"]>=70]) / total_students * 100
reading_pass = len(student_data[student_data["reading_score"]>=70]) / total_students * 100

#Create a dataframe to hold the above results
district_summary = pd.DataFrame({
    "Total Schools": total_schools,
    "Total Students": total_students,
    "Total Budget": total_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    "Average Overall Score": avg_overall,
    "% Passing Math": math_pass,
    "% Passing Reading": reading_pass,
    "% Passing Math": math_pass
}, index=[0])

#Give the displayed data cleaner formatting
district_summary["Total Students"] = district_summary["Total Students"].apply(lambda x: f"{x:,}")
district_summary["Total Budget"] = district_summary["Total Budget"].apply(lambda x: f"${x:,.2f}")
district_summary["Average Math Score"] = district_summary["Average Math Score"].apply(lambda x: f"{x:,.2f}")
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].apply(lambda x: f"{x:,.2f}")
district_summary["Average Overall Score"] = district_summary["Average Overall Score"].apply(lambda x: f"{x:,.2f}")
district_summary["% Passing Math"] = district_summary["% Passing Math"].apply(lambda x: f"{x:,.2f}%")
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].apply(lambda x: f"{x:,.2f}%")

#Display dataframe
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Group by school name
school_data_grouped = school_data_complete.groupby("school_name")
student_count = school_data_grouped["math_score"].count()

#Filter for students who scored higher than 70, group by school name, obtain counts
pass_math = student_data[student_data["math_score"]>=70].groupby("school_name")
pass_math = pass_math["math_score"].count()
pass_reading = student_data[student_data["reading_score"]>=70].groupby("school_name")
pass_reading = pass_reading["reading_score"].count()

#Begin dataframe, set index as school name
school_summary = pd.DataFrame({
    "School Name": school_data["school_name"],
    "School Type": school_data["type"],
    "Total Students": school_data["size"].apply(lambda x: f"{x:,}"),
    "Total School Budget": school_data["budget"].apply(lambda x: f"${x:,}"),
    "Per Student Budget": (school_data["budget"] / school_data["size"]).apply(lambda x: f"${x:.0f}")
}).set_index("School Name")

#Perform calculations to obtain mean scores and % of students who passed
school_math_avg = school_data_grouped["math_score"].mean()
school_reading_avg = school_data_grouped["reading_score"].mean()
percent_pass_math = pass_math / student_count * 100
percent_pass_reading = pass_reading / student_count * 100

#Format data
school_summary["Average Math Score"] = school_math_avg.apply(lambda x: f"{x:.2f}")
school_summary["Average Reading Score"] = school_reading_avg.apply(lambda x: f"{x:.2f}")
school_summary["% Passing Math"] = percent_pass_math.apply(lambda x: f"{x:.2f}%")
school_summary["% Passing Reading"] = percent_pass_reading.apply(lambda x: f"{x:.2f}%")
school_summary["% Overall Passing Rate"] = ((percent_pass_reading + percent_pass_math) / 2).apply(lambda x: f"{x:.2f}%")

school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 584, 615, 645, 10000]
group_names = ["<$585", "585-615", "616-645", "$646+"]
```

```python
#create new dataframe, make new column to show budget/student
df_school_spending = school_data_complete
df_school_spending["Per Student Budget"] = df_school_spending["budget"] / df_school_spending["size"]

#Use cut to create a new column, group by this new column, obtain count
df_school_spending["Spending Ranges(Per Student)"] = pd.cut(df_school_spending["Per Student Budget"], spending_bins, labels = group_names)
df_school_spending_grouped = df_school_spending.groupby("Spending Ranges(Per Student)")
total_school_spending = df_school_spending_grouped["math_score"].count()

#Filter for students who scored higher than 70, group by school name, obtain counts
pass_math_ss = df_school_spending[df_school_spending["math_score"]>=70].groupby("Spending Ranges(Per Student)")
pass_math_ss = pass_math_ss["math_score"].count()
pass_reading_ss = df_school_spending[df_school_spending["reading_score"]>=70].groupby("Spending Ranges(Per Student)")
pass_reading_ss = pass_reading_ss["reading_score"].count()

#Calculate
school_spending_math = df_school_spending_grouped["math_score"].mean()
school_spending_reading = df_school_spending_grouped["reading_score"].mean()
per_passing_math = pass_math_ss / total_school_spending * 100
per_passing_reading = pass_reading_ss / total_school_spending * 100

#Create data frame
scores_by_school_spending = pd.DataFrame({
    "Average Math Score": school_spending_math.apply(lambda x: f"{x:.2f}"),
    "Average Reading Score": school_spending_reading.apply(lambda x: f"{x:.2f}"),
    "% Passing Math": per_passing_math.apply(lambda x: f"{x:.2f}%"),
    "% Passing Reading": per_passing_reading.apply(lambda x: f"{x:.2f}%"),
    "% Overall Passing Rate": ((per_passing_math + per_passing_reading) / 2).apply(lambda x: f"{x:.2f}%")
})

scores_by_school_spending
```
